```python
from scipy.stats import zscore
```

```python
z_scores = zscore(df.select_dtypes(include=['float64', 'int64']))
```

```python
outliers_in_columns = np.abs(z_scores) > 3
```

```python
outlier_columns = df.select_dtypes(include=['float64', 'int64']).columns[np.any(outliers_in_columns, axis=0)]
```

```python
outlier_columns
```

```python
def remove_outliers(Df, col, k):
    mean = Df[col].mean()  # Calculate mean
    sd = Df[col].std()  # Calculate standard deviation
    # Create a list of values that are within k standard deviations from the mean
    final_list = [x for x in Df[col] if (mean - k * sd) < x < (mean + k * sd)]
    Df_cleaned = Df.loc[Df[col].isin(final_list)]  # Filter rows that have no outliers
    print(f"Outliers removed from {col}: {Df.shape[0] - Df_cleaned.shape[0]}")
    return Df_cleaned  # Return the cleaned DataFrame

```

```python
df=remove_outliers(df,'normalized_losses',3)
print(df.shape)
```
